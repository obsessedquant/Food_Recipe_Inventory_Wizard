# allrecipes.com Scraper

```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as ureq
import re
import csv
```

```python
my_url = input('Copy and past your allrecipes.com url here: ')
# my_url = 'https://www.allrecipes.com/recipe/10141/outrageous-chocolate-chip-cookies/'
# my_url = 'https://www.allrecipes.com/recipe/20185/virginas-tuna-salad/'

# Opening connection, grabbing the page
uclient = ureq(my_url)
page_html = uclient.read()
uclient.close()

# HTML parsing
page_soup = soup(page_html, "html.parser")

print('\nHTML parsing completed.')
```

```python
# First table variables filled
title = page_soup.findAll("h1",{"class":"headline heading-content elementFont__display"})[0].text
print(f'Title of this recipe is: {title}')

total_mins = page_soup.findAll("div",{"class":"recipe-meta-item-body"})[-3].text
total_mins = total_mins.split()
if len(total_mins) > 2:
    total_mins = float(total_mins[0]) * 60 + float(total_mins[2])
else:
    total_mins = float(total_mins[0])
print(f'Total minutes needed to make this recipe is: {total_mins}')

default_servings = page_soup.findAll("div",{"class":"recipe-meta-item-body"})[-2].text
default_servings = float(default_servings)
print(f'Default servings for this recipe are: {default_servings}')

# Grabs each product
containers = page_soup.findAll("span",{"class":"ingredients-item-name"})
# len(containers) # find total number of containers
```
