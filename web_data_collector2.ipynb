```python
from bs4 import BeautifulSoup
import requests

source = requests.get('http://coreyms.com').text

soup = BeautifulSoup(source, 'lxml')

for article in soup.find_all('article'): # change from find to find_all for looping
    headline = article.h2.a.text
    print(headline)

    summary = article.find('div', class_='entry-content').p.text
    print(summary)

    try: # this try and except block is used in the case that a YouTube video is not provided on the web page
        vid_src = article.find('iframe', class_='youtube-player')['src'] # gives you src dictionary attribute for iframe

        vid_id = vid_src.split('/')[4] # specifying that we want the 4th index of the list
        vid_id = vid_id.split('?')[0] # splitting by ? and taking the 0th index

        yt_link = f'https://youtube.com/watch?v={vid_id}'
        print(yt_link)
    except:
        print("An exception occurred")
    
    print()
```

```python
from bs4 import BeautifulSoup
import requests
import csv

source = requests.get('http://coreyms.com').text

soup = BeautifulSoup(source, 'lxml')

csv_file = open('cms_scrape.csv', 'w')

csv_writer = csv.writer(csv_file)
csv_writer.writerow(['headline','summary','video_link'])

for article in soup.find_all('article'): # change from find to find_all for looping
    headline = article.h2.a.text
    print(headline)

    summary = article.find('div', class_='entry-content').p.text
    print(summary)

    try: # this try and except block is used in the case that a YouTube video is not provided on the web page
        vid_src = article.find('iframe', class_='youtube-player')['src'] # gives you src dictionary attribute for iframe

        vid_id = vid_src.split('/')[4] # specifying that we want the 4th index of the list
        vid_id = vid_id.split('?')[0] # splitting by ? and taking the 0th index

        yt_link = f'https://youtube.com/watch?v={vid_id}'
        print(yt_link)
    except:
        print("An exception occurred")
    
    print()
    
    csv_writer.writerow([headline,summary,yt_link])

csv_file.close()
```

```python
import re

ini_string = input('Enter a String: ')

# code to find letters in a string
res1 = " ".join(re.split("[^a-zA-Z]*", ini_string))

# code to find numbers in a string
res2 = " ".join(re.split("[^0-9]*", ini_string))

# Printing result
print("First string result: ", str(res1))
print("Second string result: ", str(res2))
```
